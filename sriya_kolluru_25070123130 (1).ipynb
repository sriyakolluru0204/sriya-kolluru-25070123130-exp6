```python
#.6from datetime import datetime, timedelta


# Get the initial date from user input
date_str = input("Enter a date (YYYY-MM-DD): ")
initial_date = datetime.strptime(date_str, "%Y-%m-%d")

# Get the number of days to increment
days_to_add = int(input("Enter the number of days to increment: "))

# Increment the date
new_date = initial_date + timedelta(days=days_to_add)

print(f"Original date: {initial_date.strftime('%Y-%m-%d')}")
print(f"New date after incrementing by {days_to_add} days: {new_date.strftime('%Y-%m-%d')}")
```

```python
'''Problem Statement:
Write a Python program to calculate the gross salary of an employee
if basic  10000: HRA <=20%, DA =80%
if basic  20000: HRA <=25%, DA =90%
Otherwise:HRA
```

```python
basic = float(input("Enter basic salary: "))

if basic <= 10000:
    hra = basic * 0.20
    da = basic * 0.80
elif basic <= 20000:
    hra = basic * 0.25
    da = basic * 0.90
else:
    hra = basic * 0.30
    da = basic * 0.95

gross_salary = basic + hra + da

print("HRA:", hra)
print("DA:", da)
print("Gross Salary:", gross_salary)
```

Up to ₹2,50,000 : No tax

₹2,50,001 - ₹5,00,000 : 5%

₹5,00,001 - ₹10,00,000 : 20%

Above ₹10,00,000 : 30%


income = float(input("Enter annual income: "))

```python
income = float(input("Enter annual income: "))

if income <= 250000:
    tax = 0

elif income <= 500000:
    tax = (income - 250000) * 0.05

elif income <= 1000000:
    tax = (250000 * 0.05) + (income - 500000) * 0.20

else:
    tax = (250000 * 0.05) + (500000 * 0.20) + (income - 1000000) * 0.30

print("Income Tax =", tax)
```
